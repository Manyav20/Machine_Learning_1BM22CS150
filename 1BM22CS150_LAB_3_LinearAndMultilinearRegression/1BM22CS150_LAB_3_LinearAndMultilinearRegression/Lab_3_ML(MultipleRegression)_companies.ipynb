```python
import pandas as pd
df = pd.read_csv('1000_Companies.csv')
df.head()
```

```python

import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder

X = df.iloc[:, :-1]
y = df.iloc[:, -1]

encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
encoded_state = encoder.fit_transform(X[['State']])
encoded_df = pd.DataFrame(encoded_state, columns=encoder.get_feature_names_out(['State']))
X = X.drop('State', axis=1)
X = pd.concat([X, encoded_df], axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

model = LinearRegression()
model.fit(X_train, y_train)

new_data = pd.DataFrame({
    'R&D Spend': [91694.48],
    'Administration': [515841.3],
    'Marketing Spend': [11931.24],
    'State_California': [0],
    'State_Florida': [1],
    'State_New York': [0]
})

new_data = new_data[X_train.columns]

predicted_profit = model.predict(new_data)

print(f"Predicted profit: ${predicted_profit[0]:.2f}")
```
