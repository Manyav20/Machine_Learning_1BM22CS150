```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Function to convert experience words to numbers
def convert_to_numeric(value):
    word_to_num = {
        "zero": 0, "one": 1, "two": 2, "three": 3, "four": 4,
        "five": 5, "six": 6, "seven": 7, "eight": 8, "nine": 9,
        "ten": 10, "eleven": 11, "twelve": 12
    }
    return word_to_num.get(value.lower(), value) if isinstance(value, str) else value

# Load dataset
file_path = "hiring.csv"  # Ensure the file is in the same directory or provide the full path
df = pd.read_csv(file_path)

# Convert experience column to numeric
df.iloc[:, 0] = df.iloc[:, 0].apply(convert_to_numeric)

# Handle missing values (if any)
imputer = SimpleImputer(strategy='mean')
df.iloc[:, :] = imputer.fit_transform(df)

# Feature scaling
scaler = StandardScaler()
X = df.iloc[:, :-1]  # Experience, Test Score, Interview Score
y = df.iloc[:, -1]   # Salary
X_scaled = scaler.fit_transform(X)

# Train the model
model = LinearRegression()
model.fit(X_scaled, y)

# Predict salaries for given candidates
candidates = np.array([[2, 9, 6], [12, 10, 10]])
candidates_scaled = scaler.transform(candidates)
predicted_salaries = model.predict(candidates_scaled)

# Display results
for i, salary in enumerate(predicted_salaries):
    print(f"Predicted salary for candidate {i+1}: ${salary:.2f}")
```
